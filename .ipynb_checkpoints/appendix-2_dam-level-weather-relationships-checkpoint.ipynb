There are a few reasons why I think the data, as it stands, may not carry enough signal:
+ Consumption data is missing. While this is almost certainly correlated to weather in some way, it is probably not captured by weather to a large enough extent.
+ More weather data is needed. The Theewaterskloof dam only has one weather station. The weather in the surrounding areas, such as rain on the mountains, might add valuable information. + The relationship between the dam levels, geography, and meteorology are complex. For instance, how water-logged the surrounding land is could affect how much precipitation makes its way into the dam. As I mentioned above, it's important to note that this is time-series data and there are particular techniques that can be employed to analyse them. In a subsequent notebook, I might do such an analysis but for now I simply explore the relationship between *changes* in the dam level and weather variables.

```python
%matplotlib inline

import pandas as pd
import numpy as np
import datetime
import seaborn as sns
from matplotlib import pyplot as plt

# Run all nodes interactively
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
def filtered_heatmap(data, y, k = 10, largest = True):
    data = data.dropna(axis=0, how='any')
    corrmat = data.corr()
    if not largest:
        cols = corrmat.nsmallest(k, y)[y].index
    else:
        cols = corrmat.nlargest(k, y)[y].index
    if y not in cols:
        cols = cols.insert(0, y)
    cm = np.corrcoef(data[cols].values.T)
    
    print(cols)
    fig = plt.figure(figsize = (12,12))
    sns.set(font_scale = 1.25)
    hm = sns.heatmap(cm, cbar = True, annot = True, square = True, fmt = '.2f',
                     annot_kws = {'size': 10}, yticklabels = cols.values, xticklabels = cols.values)
    plt.show()

def multi_scatter(df, var_list, y_var, fit_line = False):
    df = df.dropna(axis = 0, how = 'any')
    y = df[y_var]
    for var in var_list:
        x = df[var]
        fig, ax = plt.subplots()
        ax.scatter(x, y)
        ax.set_xlabel(var, fontsize = 15)
        ax.set_ylabel(y_var, fontsize = 15)
        if fit_line:
            fit = np.polyfit(x, y, deg=1)
            ax.plot(x, fit[0] * x + fit[1], color='red')
        plt.show()    
```

```python
data = pd.read_csv('data/processed/Theewaterskloof-weather-and-dam-levels-clean-20120101-20171206.csv')
data['date'] = pd.to_datetime(data['date'])
data['diff_storage_ml'] = data['storage_ml'].diff(periods = 1)  # Get *change* in dam levels
data = data.drop('storage_ml', axis = 1)
data.index = data['date']
data.head()
```

Below are correlation matrices with the highest and lowest correlation coefficients with `diff_storage_ml`.

The variable with the greatest correlation is `heathingdegreedays` but with a coefficient of 0.38, this isn't very high. `meantempm` is the variable most negatively correlated, with a coefficient of -0.35, also weak.

We can also see that some of the variables are correlated with each other - the minimum and mean pressure, for example is extremely highly correlated, with a correlation coefficient of 0.95.

```python
print('Highest correlation coefficients:')
print()
filtered_heatmap(data, 'diff_storage_ml', 10)
print('Lowest correlation coefficients:')
print()
filtered_heatmap(data, 'diff_storage_ml', 9, largest = False)
```

The distribution plots below give us a sense of what the data looks like. Most don't seem to be well-behaved normal distributions (`minhumidity` is an exception). Some variables are discrete, like `maxvisim` and `rain` and some others are bimodal, like `gdegreedays` and `maxtempm`. These would need to be taken into account when modelling, particulary if employing linear regression. This shouldn't affect the correlation coefficients.

```python
# Only include numeric columns
num_var = data.select_dtypes(include=[np.number]).columns
data_num = data[num_var]
# Melt to long format
data_num_long = pd.melt(data_num.reset_index(), id_vars = 'date')
data_num_long.head()

grid = sns.FacetGrid(data_num_long, col = "variable", col_wrap = 3, sharex = False, sharey = False)
grid = grid.map(sns.distplot, "value")
```

Below are scatter-plots of the variables most highly (postively and negatively) correlated with `diff_storage_ml`. They confirm what we've seen earlier - the relationships between the weather variables and changes in dam levels are very tenuous. There also appears to be heteroskedastikity present in some cases (the variance in dam-level changes increases as `maxhumidity` increases for example). This can have implications for the interpretations of the variable coefficients if estimated via linear regression but might not be a problem is all we care about is predictive accuracy. There also appears to be some outliers, which might bias the results, and these would need to be dealt with when modelling.

```python
pos_corr = ['heatingdegreedays',
       'maxpressurem', 'meanpressurem', 'minpressurem', 'rain', 'maxhumidity',
       'humidity', 'minhumidity']
neg_corr = ['meantempm', 'gdegreedays', 'maxtempm',
       'maxdewptm', 'mintempm', 'meandewptm', 'mindewptm', 'coolingdegreedays',
       'maxwspdm']
multi_scatter(data, pos_corr, 'diff_storage_ml', fit_line = True)
multi_scatter(data, neg_corr, 'diff_storage_ml', fit_line = True)
```

It might take time for precipitation to reach the dam and there might be some cumulative effect. For instance, as the ground become wetter, it is less able to soak up additional moisture. So let's look at how the relationship changes if we add different cumulative summations to the precipitation variable. As we can see, it does make a difference, with the relationship becoming stronger, with 4 days showing the highest correlation. However, after 4 days, the relationship appears to become weaker. This suggests that precipitation from the previous 4 days has an effect on dam levels.

```python
window = {}
for i in range(1, 10):
    cumsum_precip = data['precipm'].rolling(window = i).sum()
    print ('\n', 'Rolling window:', i, '\n')
    window[i] = data['diff_storage_ml'].corr(cumsum_precip)
    plt.scatter(cumsum_precip, data['diff_storage_ml'])
    plt.xlabel('cumsum_precip')
    plt.ylabel('diff_storage_ml')
    plt.show();

roll = []
correl = []
for k, v in window.items():
    roll.append(k)
    correl.append(v)
plt.plot(roll, correl)
plt.xlabel('Rolling Window')
plt.ylabel('Correlation Coefficient')
plt.show();
```
