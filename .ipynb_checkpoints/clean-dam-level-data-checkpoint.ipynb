```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import datetime

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

sns.set_style("darkgrid")
```

```python
file_location = '/Users/jonathan/Google Drive/Data Science/Projects/Dam Levels/data/Dam-levels-update-2012-2017.csv'
data = pd.read_csv(file_location, header = [2,3,4], encoding = 'latin-1')

header_0 = [h[0] for h in data.columns[1:]]
header_1 = [h[1] for h in data.columns[1:]]
header_2 = [h[2] for h in data.columns[1:]]

for idx, h in enumerate(header_0):
    if re.search(r'^Unnamed', h) is not None:
        header_0[idx] = header_0[idx - 1]

header_0 = [h.title() for h in header_0]
header_1 = [h.lower() for h in header_1]
header_2 = [h.lower() for h in header_2]

col_names = ['date']
col_names = col_names + list(map('_'.join, zip(header_0, header_1, header_2)))

data.columns = pd.Index(col_names)
data = data.drop([col for col in data.columns if re.match(r'^Total Stored', col)], axis = 1)
```

```python
data['date'] = pd.to_datetime(data['date'])
data = data[:2161]
data.tail(20)
```

```python
bad_date_idx = data.loc[data['date'] > '2017-11-30'].index
# data.iloc[bad_date_idx]
max_bad_idx = max(bad_date_idx)
min_bad_idx = min(bad_date_idx)
data.iloc[min_bad_idx-20:min_bad_idx]['date'].tolist()  
print('problem starts here...')
data.iloc[min_bad_idx:max_bad_idx]['date'].tolist()
```

```python
(datetime.datetime(year = 2017, month = 11, day = 30) - data['date'].min()).days + 1
data.shape[0]
```

```python
dam_names = set(header_0)
dam_names = [d.title() for d in dam_names if not d.startswith('Total Stored')]

cols = data.columns.tolist()[1:]
dates = data['date']
# data.index = dates

dam_levels = pd.DataFrame()
for dam in dam_names:
    next_dam = pd.DataFrame({'date': dates,
                            'dam_name': dam})
    for col in cols:
        if re.search(dam, col):
            measure = re.sub(dam+'_', '', col)
            next_dam[measure] = data[col].replace(' ', '', regex = True).replace('#VALUE!', np.nan, regex = True)
            next_dam[measure] = next_dam[measure].astype(float)    
    dam_levels = pd.concat([dam_levels, next_dam], axis = 0)

dam_names = [dam.lower() for dam in dam_names]
```

```python
[d.title() for d in dam_names if not d.startswith('Total Stored')]
```

```python
dam_levels = dam_levels.rename(columns = {'height_(m)': 'height_m', 
                            'storage_(ml)': 'storage_ml',
                            'last year (%)': 'lastyear_%'})
```

```python
len(dam_names)
dam_names
storage_summary = dam_levels.pivot_table(index = 'dam_name', 
                       values = 'storage_ml', 
                       aggfunc = [min, max, np.std]).sort_values(('max', 'storage_ml'))
storage_summary
```

```python
fig, ax = plt.subplots(figsize=[20,10]) # [w, h]
sns.boxplot(dam_levels['dam_name'], dam_levels['storage_ml'], order = storage_summary.index)
plt.xticks(rotation=45);
```

```python
storage_ml_pivot = dam_levels.pivot_table(index = 'date', columns = 'dam_name', values = 'storage_ml')
agg_storage_ml = storage_ml_pivot.apply(np.sum, axis = 1)
max_storage_ml = agg_storage_ml.max()
agg_storage_perc = agg_storage_ml/max_storage_ml
```

```python
fig, ax = plt.subplots(figsize=[20,10]) # [w, h]
plt.plot(agg_storage_ml)
```

```python
dam_levels['year'] = dam_levels['date'].dt.year
dam_levels['month'] = dam_levels['date'].dt.strftime('%b')
dam_levels['week'] = dam_levels['date'].dt.week
dam_levels['dayofmonth'] = dam_levels['date'].dt.day
dam_levels['dayofyear'] = dam_levels['date'].dt.dayofyear
dam_levels['daymonth'] = dam_levels['dayofmonth'].apply(str) + '-' + dam_levels['month']
dam_levels.head()
```

```python
storage_ml_pivot = dam_levels.pivot_table(index = 'dayofyear', columns = 'year', values = 'storage_ml')
storage_ml_pivot.head()
```

```python
first_day_of_month = dam_levels.loc[(dam_levels['dayofmonth'] == 1) & (dam_levels['year'] == 2013), 
                                    ['dayofyear', 'daymonth']].drop_duplicates()
first_day_of_month
```

```python
fig, ax = plt.subplots(figsize=[20,10]) # [w, h]

for year in storage_ml_pivot.columns:
    plt.plot(storage_ml_pivot[year], linewidth = 5)
ax.legend(fontsize = 15)
ax.set_xlim(1, 366)
plt.xticks(first_day_of_month['dayofyear'], first_day_of_month['daymonth'], fontsize = 15f4fa771e705a34b1);
```
